```python
# Convierte la columna 'capture_rate' a un tipo numérico
df_pokemon_filt['capture_rate'] = pd.to_numeric(df_pokemon_filt['capture_rate'], errors='coerce')

# Normaliza la columna 'capture_rate'
minimo = df_pokemon_filt['capture_rate'].min()
maximo = df_pokemon_filt['capture_rate'].max()

df_pokemon_filt['capture_rate_normalizada'] = (df_pokemon_filt['capture_rate'] - minimo) / (maximo - minimo)
df_pokemon_filt['capture_rate_normalizada'] = df_pokemon_filt['capture_rate_normalizada'] * (10 - 0) + 0

df_pokemon_filt.head()
```

```python
#ORDENAMOS DEL MEJOR AL PEOR
df_pokemon_filt = df_pokemon_filt.sort_values(by=['base_total', 'capture_rate_normalizada'], ascending=[False, True])
df_pokemon_filt.head()
```

```python
def mochila(pokemon, capacidad):
    solucion = []
    valor_solucion = 0
    facilidad_solucion = 0

    # Ordena el DataFrame de Pokemon según la tasa de captura normalizada (ascendente) y el valor base total (descendente)
    pokemon = pokemon.sort_values(['capture_rate_normalizada', 'base_total'], ascending=[True, False])

    for index, mejor_pokemon in pokemon.iterrows():
        if len(solucion) == capacidad:
            break

        solucion.append(mejor_pokemon['name'])
        valor_solucion += mejor_pokemon['base_total']
        facilidad_solucion += mejor_pokemon['capture_rate_normalizada']

    return (solucion, valor_solucion, facilidad_solucion)

capacidad_mochila = 6
solucion, valor_total, facilidad_total = mochila(df_pokemon_filt, capacidad_mochila)

print("Solución:")
for p in solucion:
    print(p)
print("base_total:", valor_total)
```

```python
import random

def mochila(pokemon, cantidad_pokemon, capture_rate_objetivo, epsilon=0.1):
    solucion = []
    valor_solucion = 0
    facilidad_solucion = 0
    capture_rate_actual = 0

    for _ in range(1000):  # Limita el número de intentos a 1000 para evitar bucles infinitos
        pokemon_seleccionado = pokemon.sample(weights=pokemon['capture_rate'], n=1).iloc[0]
        nuevo_capture_rate = capture_rate_actual + pokemon_seleccionado['capture_rate']

        if len(solucion) < cantidad_pokemon - 1 or abs(nuevo_capture_rate - capture_rate_objetivo) < epsilon:
            solucion.append(pokemon_seleccionado['name'])
            valor_solucion += pokemon_seleccionado['base_total']
            facilidad_solucion += pokemon_seleccionado['capture_rate']
            capture_rate_actual = nuevo_capture_rate

        if len(solucion) == cantidad_pokemon:
            break

    return (solucion, valor_solucion, facilidad_solucion)

cantidad_pokemon = 6
capture_rate_objetivo = 10
solucion, valor_total, facilidad_total = mochila(df_pokemon_filt, cantidad_pokemon, capture_rate_objetivo)

print("Solución:")
for p in solucion:
    print(p)
print("base_total:", valor_total)
print("capture_rate:", facilidad_total)
```

```python
# Asegúrate de que el índice de df_pokemon_filt esté restablecido
df_pokemon_filt = df_pokemon_filt.reset_index(drop=True)

# Asegúrate de que el índice de df_pokemon_filt esté restablecido
df_pokemon_filt = df_pokemon_filt.reset_index(drop=True)

# Agrega una columna con valores aleatorios al DataFrame
df_pokemon_filt['random'] = np.random.rand(len(df_pokemon_filt))

# Ordena el DataFrame por capture_rate_normalizada y luego por el valor aleatorio
df_pokemon_filt = df_pokemon_filt.sort_values(['capture_rate_normalizada', 'random'], ascending=[True, True])
# Imprime solo las primeras 5 filas del DataFrame
df_pokemon_filt.head()


def buscar_soluciones(pokemon, capacidad, suma_objetivo, memo={}, solucion_actual=[], indice_actual=0, suma_actual=0):
    if len(solucion_actual) == capacidad:
        if abs(suma_actual - suma_objetivo) < 1e-6:
            return [solucion_actual]
        return []

    if indice_actual >= len(pokemon):
        return []

    if (indice_actual, suma_actual) in memo:
        return memo[(indice_actual, suma_actual)]

    soluciones = []

    solucion_con_pokemon_actual = solucion_actual + [indice_actual]
    suma_actual_con_pokemon_actual = suma_actual + pokemon.loc[indice_actual]['capture_rate_normalizada']
    soluciones.extend(buscar_soluciones(pokemon, capacidad, suma_objetivo, memo, solucion_con_pokemon_actual, indice_actual + 1, suma_actual_con_pokemon_actual))

    soluciones.extend(buscar_soluciones(pokemon, capacidad, suma_objetivo, memo, solucion_actual, indice_actual + 1, suma_actual))

    memo[(indice_actual, suma_actual)] = soluciones
    return soluciones

capacidad_mochila = 6
suma_objetivo = 10
solucion, valor_total, facilidad_total = mochila(df_pokemon_filt, capacidad_mochila, suma_objetivo)

print("Solución:")
for p in solucion:
    print(df_pokemon_filt.loc[p]['name'])
print("base_total:", valor_total)
print("facilidad_total:", facilidad_total)
```
