```python
# importamos las librerías necesarias 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

from matplotlib.pyplot import rcParams

rcParams['figure.figsize'] = 12, 6 # el primer dígito es el ancho y el segundo el alto
rcParams["font.weight"] = "bold"
rcParams["font.size"] = 10
rcParams["axes.labelweight"] = "bold"

# Cargamos el dataset
import pandas as pd
import chardet

file_path = 'C:\\Users\\rodri\\OneDrive\\Documentos\\GitHub\\Ejemplo_mochila_pokemon\\pokemon.csv'

with open(file_path, "rb") as f:
    result = chardet.detect(f.read())

# Leer el archivo CSV utilizando la codificación detectada
df_pokemon = pd.read_csv(file_path, encoding=result["encoding"], on_bad_lines="skip")

# Mostrar las primeras filas del DataFrame
df_pokemon.head()
#RR
```

```python
#normalizamos y hacemos la inversa



minimo = df_pokemon_filt['capture_rate'].min()
maximo = df_pokemon_filt['capture_rate'].max()

df_pokemon_filt['capture_rate_normalizada'] = (df_pokemon_filt['capture_rate'] -minimo) / (maximo - minimo)
df_pokemon_filt['capture_rate_normalizada'] = df_pokemon_filt['capture_rate_normalizada'] * (10 - 0) + 0

df_pokemon_filt.head()


```

```python
#ORDENAMOS DEL MEJOR AL PEOR
df_pokemon_filt = df_pokemon_filt.sort_values(by=['base_total', 'capture_rate_normalizada'], ascending=[False, True])
df_pokemon_filt.head()
```

```python
def mochila(pokemon, capacidad):
    solucion = []
    valor_solucion = 0
    facilidad_solucion = 0

    # Ordena el DataFrame de Pokemon según la tasa de captura normalizada (ascendente) y el valor base total (descendente)
    pokemon = pokemon.sort_values(['capture_rate_normalizada', 'base_total'], ascending=[True, False])

    for index, mejor_pokemon in pokemon.iterrows():
        if len(solucion) == capacidad:
            break

        solucion.append(mejor_pokemon['name'])
        valor_solucion += mejor_pokemon['base_total']
        facilidad_solucion += mejor_pokemon['capture_rate']

    return (solucion, valor_solucion, facilidad_solucion)

capacidad_mochila = 6
solucion, valor_total, facilidad_total = mochila(df_pokemon_filt, capacidad_mochila)

print("Solución:")
for p in solucion:
    print(p)
print("base_total:", valor_total)
```
